```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('1_feature_engineered.csv')
```

```python
data.head()
```

```python
data.drop(data.columns[0], axis=1, inplace=True)
```

```python
data.columns
```

```python
Xcols = ['price value', 'total open hours', 'checkin per hour',
       'Alcohol', 'BikeParking', 'BusinessAcceptsCreditCards', 'Caters',
       'GoodForDancing', 'GoodForKids', 'DogsAllowed', 'dinner',
       'HappyHour', 'HasTV', 'Music', 'OutdoorSeating', 'RestaurantsDelivery',
       'RestaurantsGoodForGroups', 'RestaurantsReservations',
       'RestaurantsTakeOut', 'Smoking',
       'WheelchairAccessible', 'WiFi', 'casual', 'classy', 'divey', 'hipster',
       'intimate', 'romantic', 'touristy', 'trendy', 'upscale', 'breakfast',
       'brunch', 'dessert', 'latenight', 'lunch', 'garage', 'lot',
       'street', 'valet', 'validated', 'average', '0', 'loud', 'quiet',
       'very_loud', 'Summerlin', 'Sunrise', 'Eastside', 'Southeast',
       'Spring Valley', 'Northwest', 'The Strip', 'Westside', 'Downtown',
       'Chinatown', 'Centennial', 'Southwest', 'The Lakes', 'South Summerlin',
       'University', 'Anthem', 'dressy', 'formal', 'DriveThru']
treatment = 'RestaurantsTableService'
```

```python
def print_nonnumeric_values(x, column_name):
    try:
        float(x)
    except ValueError:
        print(column_name + ': ' + x)


for column in Xcols:
    data[column].apply(lambda x: print_nonnumeric_values(x, column))


for column in ['Smoking', 'WiFi']:
    data[column] = pd.to_numeric(data[column], errors='coerce')
data['Smoking'] = data['Smoking'].fillna(0)
data['WiFi'] = data['WiFi'].fillna(1)
```

```python
for col in Xcols+['stars', treatment]:
  data[col] = pd.to_numeric(data[col], errors='raise')
```

```python
Y = data['stars'].values
X = data[Xcols].values
D = data[treatment].values
```

```python
from causalinference import CausalModel
```

```python
ols = CausalModel(Y, D, X)
ols.est_via_ols()
print(ols.estimates)
```

```python
matching = CausalModel(Y, D, X)
matching.est_via_matching()
print(matching.estimates)
```
